# Plotting

### Option 1
Use `pandas.DataFrame.plot` to automatically interface with matplotlib

```python
def plot_pandas(df, title, xlabel, ylabel, loglog):
    plot = df.plot(kind="line", 
        marker='o', 
        title=title, 
        xlabel=xlabel,
        ylabel=ylabel, 
        grid=True,
        loglog=loglog
    )
    return plot
```
